```python
import random
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import json

def plot_image_with_data(img_path, table_data):
    # Display the image
    img = mpimg.imread(img_path)
    plt.figure(figsize=(8, 6))  # Decent size for the image
    plt.imshow(img)
    plt.axis('off')  # Hide axes
    plt.show()

    # Print the table data as a string
    print("Data Table:")
    formatted_table = table_data.replace(' & ', '\n').replace('|', '\t')
    print(formatted_table)

def plot_random_images(data_json, image_base_path):
    num_images = min(4, len(data_json))
    
    # Randomly select 'num_images' unique indices
    selected_indices = random.sample(range(len(data_json)), num_images)
    
    for index in selected_indices:
        img_data = data_json[index]
        img_path = f"{image_base_path}/{img_data['img_id']}"
        table_data = img_data['table']
        plot_image_with_data(img_path, table_data)
```

```python
images_base_path = "../../datasets/unichart-pretrain-images/content/images"
json_path = "./filtered_unichart_pretrain_datatable.json"
```

```python
with open(json_path, 'r') as file:
    data_json = json.load(file)

plot_random_images(data_json, images_base_path)
```
